Задаём порог активации нейрона

```
class Neuron():
    def __init__(self, bias):
        self.bias = bias
        print('Bias: {}'.format(bias))

    def forward(self,x1,x2):
        _sum = x1+x2
        return _sum>self.bias
```

```
neuron = Neuron(0.5)
```

```
neuron.forward(x1=0.2, x2=0.4)
```

Добавляем веса

```
class Neuron():
    def __init__(self, bias, w1, w2):
        self.bias = bias
        self.w1 = w1
        self.w2 = w2

    def forward(self,x1,x2):
        _sum = self.w1*x1+self.w2*x2
        return _sum>self.bias
```

# Вектор

Запись с помощью вектора весов


```
import numpy as np #загружаем библиотеку для работы с матрицами
```

```
class Neuron():
    def __init__(self, weights, bias):
        self.bias = bias
        self.weights = weights

    def forward(self,x):
        _sum = np.dot(self.weights,x)
        return int(_sum-self.bias>0)
```

```
w = np.array([0.2, 0.8]) #размеры 2x1

neuron = Neuron(weights=w, bias=0.5)

x = np.array([1,1]) #размеры 1x2

neuron.forward(x)
```
